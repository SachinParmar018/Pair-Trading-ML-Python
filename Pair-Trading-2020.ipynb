```python
import tkinter as fileimport
from tkinter import filedialog
import pandas as pd

root= fileimport.Tk()

canvas1 = fileimport.Canvas(root, width = 200, height = 200, bg = 'lightsteelblue2', relief = 'raised')
canvas1.pack()

def getCSV ():
    global df2
    
    import_file_path = filedialog.askopenfilename()
    df2 = pd.read_csv (import_file_path)
    #print (df)
    
browseButton_CSV = fileimport.Button(text="      Import CSV File 2    ", command=getCSV, bg='green', fg='white', font=('helvetica', 12, 'bold'))
canvas1.create_window(100, 100, window=browseButton_CSV)

root.mainloop()

df2=df2.drop(["Open","High","Low","Qty","PeriodNo","OADate"],axis=1)
df2

#Join on Time (Merge)
df_ratio=pd.merge(df,df2,on="Time",how="inner",indicator=False)
df_ratio=df_ratio.rename(columns={"Close_x":"asian"})
df_ratio=df_ratio.rename(columns={"Close_y":"berg"})
df_ratio["Time"]=pd.to_datetime(df_ratio["Time"],infer_datetime_format=True)
df_ratio=df_ratio.sort_values(['Time'],ascending=True)
df_ratio["Ratio"]=df_ratio["asian"]/df_ratio["berg"]
#df_ratio=df_ratio.set_index("Time")
df_ratio1=pd.DataFrame(df_ratio)
df_ratio
```

```python
year=[]
for x in df_ratio['Time']:
    year.append(x.year)
df_ratio["Year"]=year
```

```python
user_yr=input("Enter year on which wants to Trade:")
user_yr=int(user_yr)
temp_yr=user_yr-1
```

```python
df=df_ratio[(df_ratio['Year']>=user_yr)]
df=df.set_index("Time")
mean_yr=df_ratio[(df_ratio['Year']==temp_yr)]
temp=df_ratio[(df_ratio['Year']==user_yr)]
final=pd.DataFrame(columns=["Date","Asian","Berg","Ratio","Year","Status","Type","Amt"])
Stop="False"
```
